```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("/content/train.csv")
df.head(10)
```

```python
#for categorical data
sns.countplot(df["Embarked"])
```

```python
df['Sex'].value_counts().plot(kind='pie', autopct="%.2f")#pie chart can be used for showing % distribution
```

```python
df["Survived"].value_counts().plot(kind="pie",autopct="%.2f")
```

```python
survived_men = df[(df['Sex'] == 'male') & (df['Survived'] == 1)].shape[0]
not_survived_men = df[(df['Sex'] == 'male') & (df['Survived'] == 0)].shape[0]
survived_women = df[(df['Sex'] == 'female') & (df['Survived'] == 1)].shape[0]
not_survived_women = df[(df['Sex'] == 'female') & (df['Survived'] == 0)].shape[0]

# Pie chart data
men_counts = [survived_men, not_survived_men]
women_counts = [survived_women, not_survived_women]
labels = ['Survived', 'Not Survived']

# Plot pie charts
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.pie(men_counts, labels=labels, autopct='%.2f%%', startangle=140)
ax1.set_title('Survival Rate of Men')

ax2.pie(women_counts, labels=labels, autopct='%.2f%%', startangle=140)
ax2.set_title('Survival Rate of Women')

plt.show()
```
