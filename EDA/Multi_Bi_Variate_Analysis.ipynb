```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
titanic = pd.read_csv('train.csv')
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
# scatter plot (numerical -numerical) data
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

```python
dia= sns.load_dataset("diamonds")
dia
```

```python
sns.scatterplot(x=dia['price'],y=dia['carat'],hue=dia['table'])#,style=tips['smoker'],size=tips['size'])
```

```python
#bar plot  (numerical - categorical)
sns.barplot(x=titanic["Pclass"],y=titanic['Age'])
```

```python
sns.barplot(x=titanic["Pclass"],y=titanic['Fare'])
```

```python
sns.barplot(x=titanic["Pclass"],y=titanic['Age'],hue=titanic["Sex"])
```

```python
# box plot (Numerical- Categorical)
sns.boxplot(x=titanic["Sex"],y=titanic['Age'],hue=titanic["Survived"])
```

```python
#distplot (numerical - categorical)
sns.distplot(titanic[titanic["Survived"]==0]["Age"])
sns.distplot(titanic[titanic["Survived"]==1]["Age"])
```

```python
#heatmap (categorical- categorical)
sns.heatmap(pd.crosstab(titanic["Pclass"],titanic["Survived"]))   # Crosstab is used for summarizing relationship between categorical variables
```

```python
titanic.groupby("Pclass")["Survived"].mean()
```

```python
titanic.groupby("Sex")["Survived"].mean()
```

```python
titanic.groupby("Embarked")["Survived"].mean()
```

```python
#cluster map ( categorical - categorical)
sns.clustermap(pd.crosstab(titanic["Parch"],titanic["Survived"]))
```

```python
#pair plot (collection of scatter plot) used for (numerical - numerical ) data
tips.head()
```

```python
sns.pairplot(tips,hue="time") #adding hue = "time" can provide more focoused charts on the basis of a categorical data
```
