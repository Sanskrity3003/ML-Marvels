```python
batman= pd.read_csv("/content/sharma-kohli.csv")
batman
```

```python
#ploting the graph for one person
plt.plot(batman['index'],batman["V Kohli"])
```

```python
#plotting graph for multiple people in one grsaph
plt.plot(batman['index'],batman["V Kohli"])
plt.plot(batman['index'],batman["RG Sharma"])
```

```python
#labling the graph
plt.plot(batman['index'],batman["V Kohli"])
plt.plot(batman['index'],batman["RG Sharma"])
plt.xlabel('Years')
plt.ylabel("Scores")
plt.title("Scores of 2 batsman over the years")
```

```python
#changing the colour of lines
plt.plot(batman['index'],batman["V Kohli"],color="red")
plt.plot(batman['index'],batman["RG Sharma"],color="yellow")
plt.xlabel('Years')
plt.ylabel("Scores")
plt.title("Scores of 2 batsman over the years")
# NOTE :- You can also use hex codes for innumerous choice colours
```

```python
#Changing the style and width of the lines
plt.plot(batman['index'],batman["V Kohli"], linestyle="dashed", linewidth=4)
plt.plot(batman['index'],batman["RG Sharma"],linestyle="dotted")
plt.xlabel('Years')
plt.ylabel("Scores")
plt.title("Scores of 2 batsman over the years")
# some more examples of style can be :- dashed , dashdot
```

```python
# adding markers and manipulating their sizes
plt.plot(batman['index'],batman["V Kohli"], linestyle="dashed", linewidth=2, marker ='D',markersize=10)
plt.plot(batman['index'],batman["RG Sharma"],linestyle="dotted",linewidth=2, marker ='o',markersize=10)
plt.xlabel('Years')
plt.ylabel("Scores")
plt.title("Scores of 2 batsman over the years")
```

```python
# Legend = labling the lines
plt.plot(batman['index'],batman["V Kohli"], linestyle="dashed", linewidth=2, marker ='D',markersize=10,label ="Virat")
plt.plot(batman['index'],batman["RG Sharma"],linestyle="dotted",linewidth=2, marker ='o',markersize=10,label ="Rg Sharma")
plt.xlabel('Years')
plt.ylabel("Scores")
plt.title("Scores of 2 batsman over the years")
plt.legend()
```
